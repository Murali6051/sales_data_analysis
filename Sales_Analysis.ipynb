# Importing all libraries

```python
# importing pandas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data Set Links

apr = 'https://raw.githubusercontent.com/ravi-varma-ankam/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_April_2019.csv'
aug = 'https://raw.githubusercontent.com/ravi-varma-ankam/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_August_2019.csv'
dec = 'https://raw.githubusercontent.com/ravi-varma-ankam/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_December_2019.csv'
feb = 'https://raw.githubusercontent.com/ravi-varma-ankam/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_February_2019.csv'
jan = 'https://raw.githubusercontent.com/ravi-varma-ankam/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_January_2019.csv'
jul = 'https://raw.githubusercontent.com/ravi-varma-ankam/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_July_2019.csv'
jun = 'https://raw.githubusercontent.com/ravi-varma-ankam/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_June_2019.csv'
mar = 'https://raw.githubusercontent.com/ravi-varma-ankam/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_March_2019.csv'
may = 'https://raw.githubusercontent.com/ravi-varma-ankam/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_May_2019.csv'
nov = 'https://raw.githubusercontent.com/ravi-varma-ankam/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_November_2019.csv'
oct = 'https://raw.githubusercontent.com/ravi-varma-ankam/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_October_2019.csv'
sep = 'https://raw.githubusercontent.com/ravi-varma-ankam/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_September_2019.csv'
```

```python
# storing all links in a list

all_files = [apr, aug, dec, feb, jan, jul, jun, mar, may, nov, oct, sep]
df = pd.DataFrame()
```

```python
# Joining all data in a single Dataframe

for link in all_files:
  data = pd.read_csv(link)
  df = pd.concat([df, data])
df.shape
```

#Cleaning Data

```python
# Adding new month column in the data

df['Month'] = df['Order Date'].str[0:2]
df.head(5)
```

```python
# Counting all NaN values

df.isnull().sum().sum()
```

```python
# Finding values in a month column

df['Month'].value_counts()
```

```python
# displaying NaN rows in a data

df[df.isnull().any(axis = 1)]
```

```python
# droping all NaN rows
df = df.dropna()
df
```

```python
# finding 'Or' and deleting it
df = df.drop(df[df['Month'] == 'Or'].index)
df
```

```python
df.Month.value_counts()
```

```python
df
```

```python
# Converting string to integer

df['Month'] = pd.to_numeric(df['Month'])
```

```python
df
```

```python
df['Month'][0]
```

```python
# Add a sales column

df.head(3)
```

```python
# converting string to int

df['Price Each'] = pd.to_numeric(df['Price Each'])
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
```

```python
# Adding Sales Column

df['Sales'] = df['Price Each'] * df['Quantity Ordered']
```

```python
df.head(5)
```

```python
# Best Month for Sales

sale_results = df.groupby('Month').sum()['Sales']
sale_results
```

```python
months = range(1,13)
plt.bar(months, sale_results)
plt.show()
```

# What city have a highest number of sales

```python
# What city have a highest number of sales
df.head(10)
```

```python
# Adding new column city
df['City'] = df['Purchase Address'].str.split(',').str[1]
```

```python
df.head(10)
```

```python
# showing results of sales of cities
results  = df.groupby('City').sum()['Sales']
results
```

```python
# Showing data in a graph
cities = df['City'].unique()

cities = [city for city, data in df.groupby('City')]

plt.bar(cities, results)
plt.xticks(cities, rotation = 'vertical', size = 10)
plt.ylabel('Sales in USD $')
plt.xlabel('City Name')
plt.show()
```

# What time should we display advertisements to maximize sales?


```python
# What time should we display advertisements to maximize sales?
df.head(5)
```

```python
# Adding time Column to data
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Hour'] = df['Order Date'].dt.hour
df['Minute'] = df['Order Date'].dt.minute
df.head(5)
```

```python
df.head(5)
```

```python
hours = [hour for hour, data in df.groupby('Hour')]
plt.plot(hours, df.groupby(['Hour']).count())
plt.xticks(hours)
plt.grid()
plt.show()
```

# What product sold the most?

```python
p_group = df.groupby('Product')
q_ordered = df.groupby('Product').sum()['Quantity Ordered']

products = [product for product, df in p_group]

plt.bar(products, q_ordered)
plt.ylabel('Quantity ordered')
plt.xlabel('Product')
plt.xticks(products, rotation = 'vertical', size = 10)

plt.show()
```
